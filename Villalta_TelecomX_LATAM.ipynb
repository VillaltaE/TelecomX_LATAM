```python
#Normalizar datos
df_phone = pd.json_normalize(df['phone'])
df_internet = pd.json_normalize(df['internet'])
df_account = pd.json_normalize(df['account'])
```

```python
#Obtenga los recuentos de cada categorÃ­a de persona mayor entre los clientes que se dieron de baja
churned_senior_citizen_counts = yes_churn.groupby('SeniorCitizen')['SeniorCitizen'].count()

# Obtenga los recuentos de cada categorÃ­a de persona mayor entre los clientes que no se han dado de baja
non_churned_senior_citizen_counts = no_churn.groupby('SeniorCitizen')['SeniorCitizen'].count()

# Asigne los valores binarios a etiquetas comprensibles para el eje x
labels = ['No Senior Citizen', 'Senior Citizen']

# Obtenga los recuentos para trazar, lo que garantiza que manejemos los casos en los que pueda faltar una categorÃ­a
churned_counts = [churned_senior_citizen_counts.get(0, 0), churned_senior_citizen_counts.get(1, 0)]
non_churned_counts = [non_churned_senior_citizen_counts.get(0, 0), non_churned_senior_citizen_counts.get(1, 0)]

#Crear figura y ejes
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

# Define el ancho de las barras y las posiciones en el eje x
bar_width = 0.35
x = np.arange(len(labels))

#Grafique las barras para los clientes que no se han dado de baja
rects1 = ax.bar(x - bar_width/2, non_churned_counts, bar_width, label='Usuarios actuales', color='skyblue')

# Grafique las barras para los clientes que se han dado de baja
rects2 = ax.bar(x + bar_width/2, churned_counts, bar_width, label='Usuarios evadidos', color='salmon')

# Agregue algo de texto para etiquetas, tÃ­tulo y etiquetas de marca del eje x personalizadas, etc.
ax.set_xlabel('Estatus de Adulto Mayor')
ax.set_ylabel('NÃºmero de usuarios')
ax.set_title('DistribuciÃ³n de usuarios por estatus de evasiÃ³n y estatus de adulto mayor')
ax.set_xticks(x) # Establezca las ubicaciones de las marcas de verificaciÃ³n para que estÃ©n en el medio de las dos barras
ax.set_xticklabels(labels) # Establecer las etiquetas para los ticks

# Agregar leyenda
ax.legend()

# Agregue recuentos en la parte superior de las barras
def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 puntos de desplazamiento vertical
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)


# Ajustar el diseÃ±o para evitar que las etiquetas se superpongan
fig.tight_layout()


plt.show()
```
